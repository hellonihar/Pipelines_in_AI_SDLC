This notebook is for **exploration only**. Training code lives in `scripts/train.py`.
In Level 2, the notebook is no longer the application — it's a playground for understanding the data.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("../data/customer_data.csv")
print(f"Rows: {len(df)}, Columns: {len(df.columns)}")
df.head()
```

```python
print("=== TARGET DISTRIBUTION ===")
print(df['churned'].value_counts())
print(f"Churn rate: {df['churned'].mean():.2%}")

fig, axes = plt.subplots(1, 2, figsize=(10, 4))
df['churned'].value_counts().plot(kind='bar', ax=axes[0], title='Churn Count')
df['churned'].value_counts().plot(kind='pie', ax=axes[1], autopct='%1.1f%%', title='Churn Proportion')
plt.tight_layout()
plt.show()
```

```python
print("=== NUMERIC FEATURES ===")
df.describe()
```

```python
numeric_cols = ['age', 'tenure_months', 'monthly_charges', 'total_charges',
                'avg_monthly_usage_hours', 'late_payments_last_12m']

fig, axes = plt.subplots(2, 3, figsize=(15, 10))
for i, col in enumerate(numeric_cols):
    ax = axes[i // 3][i % 3]
    ax.hist(df[col], bins=20, edgecolor='black')
    ax.set_title(col)
plt.tight_layout()
plt.show()
```

```python
cat_cols = ['contract_type', 'payment_method', 'internet_service', 'tech_support']

fig, axes = plt.subplots(2, 2, figsize=(12, 8))
for i, col in enumerate(cat_cols):
    ax = axes[i // 2][i % 2]
    crosstab = pd.crosstab(df[col], df['churned'], normalize='index')
    crosstab.plot(kind='bar', ax=ax, stacked=True, title=f'Churn Rate by {col}')
    ax.set_ylabel('Proportion')
    ax.legend(['Stay', 'Churn'])
plt.tight_layout()
plt.show()
```

```python
print("=== CORRELATION WITH TARGET ===")
corr = df[numeric_cols + ['churned']].corr()['churned'].drop('churned').sort_values()
print(corr)

plt.figure(figsize=(8, 4))
corr.plot(kind='barh', title='Correlation with Churn')
plt.tight_layout()
plt.show()
```

```python
print("=== KEY FINDINGS ===")
print("1. Churn rate: ~{:.1%}".format(df['churned'].mean()))
print("2. Month-to-month contracts have highest churn (check the stacked bars)")
print("3. Late payments correlate with churn")
print("4. Fiber optic shows higher churn than DSL")
print("5. Tech support subscribers churn less")

print("\n---")
print("Next step: train a model using: uv run python scripts/train.py")
```
